```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
X=iris.iloc[:,:-1]
Y=iris.iloc[:,-1]
print(X)
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
target=encoder.fit_transform(Y)
target
```

```python
pca = PCA(n_components=2)
pca.fit(X)
print(sum(pca.explained_variance_ratio_))
print(pca.explained_variance_ratio_)
pc=pca.fit_transform(X)
print(pc.shape)

pc_df=pd.DataFrame(
    pc,columns=['PC1','PC2']
)
pc_df["species"]=Y # agr cloumn ho ga is nam ka tuw update kary ga warna create kar kay value dal day ga.
print(pc_df)
```

```python
sns.scatterplot(data=pc_df, x="PC1", y="PC2", hue="species")
```

```python
(X_train, y_train), (X_test,y_test) = mnist.load_data()
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
plt.imshow(X_test[100])
print(y_test[100])
```

```python
x=X_test.reshape(10000,28*28)
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler  #to normalize data
standardized_data = StandardScaler().fit_transform(x)
```

```python
print(standardized_data)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(standardized_data)
principal_df = pd.DataFrame(data = principalComponents
         ,columns=['PC1','pc2'])
principal_df
```

tSNE: t_distributed shoctahastic neighbor embedding

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)

X_reduced = tsne.fit_transform(x)
```

```python
X_reduced = pd.DataFrame(X_reduced, columns=["Dim1", "Dim2"])
X_reduced["labels"]=y_test
```

```python
plt.figure()
sns.scatterplot(data=X_reduced,x="Dim1",y="Dim2", hue='labels',palette='Paired')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```
